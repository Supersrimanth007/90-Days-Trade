```python
# Extracting Data from Trade_History
def parse_trade_history(trade_history):
    if pd.isna(trade_history):
        return []  # Return an empty list or handle it as needed
    return ast.literal_eval(trade_history)

# Initialize an empty list to collect the trade histories
all_trade_histories = []

for _, row in df.iterrows():
    port_id = row['Port_IDs']
    trade_histories = parse_trade_history(row['Trade_History'])
    
    for trade in trade_histories:
        trade['Port_IDs'] = port_id  # Add the Port_IDs to the trade dictionary
        all_trade_histories.append(trade)

# If you want to convert this back to a DataFrame
trade_histories_df = pd.DataFrame(all_trade_histories)
```

```python
# creating function for max_dropdown
def calculate_max_drawdown(equity_curve):
    # Convert Series to a NumPy array for easy access
    equity_array = equity_curve.values
    peak = equity_array[0]  # Get the first value of the equity curve
    max_drawdown = 0

    for value in equity_array:
        if value > peak:
            peak = value
        drawdown = (peak - value) / peak
        if drawdown > max_drawdown:
            max_drawdown = drawdown

    return max_drawdown

```

```python
trade_histories_df['equity_curv'] = trade_histories_df.groupby('Port_IDs')['realizedProfit'].cumsum() 
```

```python
MDD = trade_histories_df.groupby(['Port_IDs'])['equity_curv'].apply(calculate_max_drawdown)
```

```python
# Merging Total_investment, Total_profit, MDD, std_profit, mean_profit, total_position, winning_pos

# Merge the first two DataFrames
df2 = pd.merge(Total_investment, Total_profit, on='Port_IDs', how='outer')

# Merge the result with MDD
df2 = pd.merge(df2, MDD, on='Port_IDs', how='outer')

# Merge with std_profit
df2 = pd.merge(df2, std_profit, on='Port_IDs', how='outer')

# Merge with mean_profit
df2 = pd.merge(df2, mean_profit, on='Port_IDs', how='outer')

# Merge with total_position
df2 = pd.merge(df2, total_position, on='Port_IDs', how='outer')

# Merge with winning_pos
df2 = pd.merge(df2, winning_pos, on='Port_IDs', how='outer')
```
